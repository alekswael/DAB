```python
def search_strings(data, patterns, context_length=100):
    """
    Search for strings in the data list that match any of the specified regex patterns,
    and return the matched string along with the surrounding context.

    Args:
        data (list): List of strings to search.
        patterns (list or str): Regex patterns to search for.
        context_length (int): Number of characters before and after the match to include in the result.

    Returns:
        list: Strings containing the match along with the context.
    """
    # Ensure patterns is a list
    if isinstance(patterns, str):
        patterns = [patterns]

    # Compile regex patterns for efficiency
    regex_patterns = [re.compile(p) for p in patterns]

    full_strings = [s for s in data if any(p.search(s) for p in regex_patterns)]

    context_strings = []
    index_col = []

    # Iterate over each string in the data
    for i in range(len(data)):
        for p in regex_patterns:
            # Find all matches in the string
            for match in p.finditer(data[i]):
                # Get the start and end positions of the match
                start, end = match.span()

                # Get the surrounding context (100 characters before and after the match)
                start_context = max(0, start - context_length)
                end_context = min(len(data[i]), end + context_length)
                context = data[i][start_context:end_context]

                # Add the context to the result
                context_strings.append(context)
                index_col.append(i)
    
    return full_strings, pd.DataFrame(context_strings, index_col)
```

```python
sensitive_keywords = ["cpr", "tlf"]

tlf_pattern = "|".join(
    [
        r"\+\d{10}",
        r"\+\d{4} \d{2} \d{2} \d{2}",
        r"\+\d{2} \d{8}",
        r"\+\d{2} \d{2} \d{2} \d{2} \d{2}",
        r"\+\d{2} \d{4} \d{4}",
        r"\d{2} \d{4} \d{4}",
        r"\d{2} \d{4}\-\d{4}",
        r"\d{8}",
        r"\d{4} \d{4}",
        r"\d{4}\-\d{4}",
        r"\d{2} \d{2} \d{2} \d{2}",
    ]
)

cpr_pattern = "|".join(
    [
        r"[0-3]\d{1}[0-1]\d{3}-\d{4}", r"[0-3]\d{1}[0-1]\d{3} \d{4}"
    ]
)

mail_pattern = r"[\w\.-]+@[\w\.-]+(?:\.[\w]+)+"
```
